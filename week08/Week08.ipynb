```python
# Import required libraries
from urllib.request import urlopen
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk

# Download required NLTK resources
nltk.download('stopwords')  # For removing common words
nltk.download('punkt_tab')      # For tokenization
```

```python
# URL for 'Les Miserables' text file from Project Gutenberg
url = 'http://www.gutenberg.org/files/135/135-0.txt'

# Fetch the raw text from the URL
book_raw = urlopen(url).read().decode('utf-8')

# Preview the first 500 characters to ensure the book content is fetched properly
print(book_raw[:500])
```

```python
# Tokenize the book's content into words
tokens = word_tokenize(book_raw)

# Display the first 10 tokens to verify
print("First 10 tokens:", tokens[:10])
```

```python
# Get the set of English stopwords
stop_words = set(stopwords.words('english'))

# Filter tokens to exclude stopwords and punctuation
filtered_tokens = [word for word in tokens if word.isalpha() and word.lower() not in stop_words]

# Display the first 10 filtered tokens to verify
print("First 10 filtered tokens:", filtered_tokens[:10])
```

```python
from collections import Counter

# Count the frequencies of each word
word_counts = Counter(filtered_tokens)

# Display the 10 most common words
print("Most common words:", word_counts.most_common(10))
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Create a word cloud
wordcloud = WordCloud(width=1000, height=700, background_color='black').generate(' '.join(filtered_tokens))

# Display the word cloud
plt.figure(figsize=(10, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
